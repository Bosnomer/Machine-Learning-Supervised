# Pandas Basics

```python
# Importing Libraries

import pandas as pd
import numpy as np

df = pd.read_pickle('country.pkl') 
# pickle is also a type of file where models, dataframes can be stored

df.head()
```

```python
# Unique values in a column

df['continent'].unique()
```

```python
# Getting values with continent as Asia

asia = df["continent"] == "Asia" # Will get all the indices and return boolean
print(asia)
print()

temp1 = df[asia] # Using indices to extract the rows
temp1.head()
```

```python
# OR

df[df["continent"]=="Asia"]
```

```python
# Getting values with Max value of year

temp = df[df['year']==df['year'].max()]
temp
```

```python
# Getting values with values less than mean of Life Expectancy

temp2=df[df["life exp"]<df["life exp"].mean()]
temp2
```

```python
# Getting values with more than one conditions

temp3 = df[(df["continent"]=="Asia") & (df["life exp"]>df["life exp"].mean())]
temp3.head()
```

```python
temp3 = df[(df["continent"]=="Asia") | (df["life exp"]<df["life exp"].mean())]
temp3.head()
```

```python
temp4 = df[(df["continent"]=="Asia") | (df["continent"]=="Asia")]
temp4.head()
```

```python
# OR

temp5 = df[df["continent"].isin(['Asia','Africa'])]
temp5.head()
```

```python
#Inverse of above

temp6 = df[~df["continent"].isin(['Asia','Africa'])]
temp6.head()
```

```python
# Sorting by conditions
# 1. Sort by year
# 2. Sort by year in descending
# 3. Sort by Population descending and NA values first


#df.sort_values(['year'])
#df.sort_values(['year'],ascending=False)
df.sort_values(['population'],ascending=False, na_position='first')
```

```python
# Sort by Year in ascending and then by Population in descending

df.sort_values(['year', 'population'],ascending=(True,False))
```

```python
?df.sort_values
```

```python
# Rename Population to pop

df = df.rename({'population':'pop'}, axis=1)
df.head()
```

```python
# Renaming Row index

df = df.rename({1001:1000}, axis=0)
df.head
```

```python
# Getting values of DataFrame by specifying Columns

df = df[["country","continent","year","pop","life exp","gdp_cap"]]
df.head()
```

```python
# Missing Values Check

df.isna() # will return Binary
```

```python
# Check if there is any Missing value in any column

df.isna().any()
```

```python
# Simply dropping NA (Not Recommended)

temp = df.dropna(axis=1)
temp
```

```python
# Drop Column if it has more than 2% of NA

temp = df.dropna(axis=1, thresh=len(df)*0.98) # drop column if it has more than 2% of na
temp
```

```python
# Count of Non Missing values in each column

df.count()
```

```python
# Creating copy of Data Frame

temp = df.copy()
print(temp.head())

print("-----")

# Filling NA values of Life expectancy of temp with 0
temp['life exp'].fillna(0, inplace=True)
print(temp.head())
```

```python
# Filling Missing values with previous row value for Life exp column

temp = df.copy()
temp['life exp'].fillna(method="ffill", inplace=True)
temp.head()
```

```python
# Filling Missing values with next row value for Life exp column

temp = df.copy()
temp['life exp'].fillna(method="bfill", inplace=True)
temp.head()
```

### Iterating through Rows: iterrows and itertuples

```python
df.iterrows()
```

```python
for index, row in df.iterrows():
    print(row)
```

```python
for index, row in df.iterrows():
    print(row['pop'])
```

```python
# Replacing values by iterrating over rows

for index, row in df.iterrows():
    if df.loc[index,'year']>=2000:
        df.loc[index, 'Century'] ="20th Century"
    else:
        df.loc[index, 'Century'] ="19th Century"
        
df.head()
```

```python
for index,(country,continent, year, pop, life, gdp_cap, Century) in df.iterrows():
    print(country, year, life)
```

```python
for rows in df.itertuples():
    print(rows)
```

```python
for (index, country,continent, year, pop, life, gdp_cap, Century) in df.itertuples():
    print(country, year, life)
```

### Comparing time in iterrows and itertuples

```python
from datetime import datetime

start = datetime.now()
for index, row in df.iterrows():
    print(row)

print()
print("Time:", datetime.now()-start)
```

```python
from datetime import datetime

start = datetime.now()
for row in df.itertuples():
    print(row)

print()
print("Time: ",datetime.now()-start)
```

### Some more Operations

```python
df = df.copy()
df.loc[df['year']>2000,'continent'] = "Asia"

df.head()
```

```python
# column1 * column2

df['total_gdp']=df['gdp_cap']*df['pop']
df.head()
```

```python
# Give mean of Numerical columns group by Year

temp = df.groupby('year').mean()
temp.head()
```

```python
# Give mean of Numerical columns group by Country

temp = df.groupby('country').mean()
temp.head()
```

```python
# Give median of columns: pop and gdp_cap group by Year

temp = df.groupby('year')['pop','gdp_cap'].median()
temp.head()
```

```python
temp = df[df['year']==df['year'].max()].groupby('continent').size()
temp
```

```python
temp = df.groupby('country')['life exp'].agg(lambda x: max(x)-min(x))
temp.head()
```

```python
temp = df.groupby(['year','continent'])['gdp_cap'].mean()
temp = temp.unstack(level=1)
temp.head(20)
```

```python
temp = df.groupby(['year','continent'])['gdp_cap'].mean()
temp = temp.unstack(level=0)
temp.head(20)
```

### Pivot Table

```python
# Pivot Table by Rows = Year and Columns = Continents and Values = Life exp

temp1 = pd.pivot_table(df, index="year",columns="continent",values = "life exp")
temp1
```

```python
# Pivot Table by Rows = Continent, Country and Columns = Year and Values = Life exp


temp1 = pd.pivot_table(df, index=["continent","country"],columns="year",values = "life exp")
temp1
```

```python
# Pivot Table by Rows = Year and Columns = Continents and Values = Sum of Life exp


temp1 = pd.pivot_table(df, index="year",columns="continent",values = "life exp", aggfunc=np.sum)
temp1
```

```python
# Pivot Table by Rows = Year and Columns = Continents and Values = Sum of Life exp and Mean of Life exp


temp1 = pd.pivot_table(df, index="year",columns="continent",values = "life exp", aggfunc=[np.sum, np.mean])
temp1
```

```python
# Pivot Table by Rows = Continents and Columns = Year and Values = Max(life)-Min(life)


temp1 = pd.pivot_table(df, index="continent",columns="year",values = "life exp", aggfunc=[lambda life: max(life)-min(life)])
temp1
```

```python
# Shifting values

temp=df.copy()
temp['life1']=temp['life exp'].shift(-4)
temp.head(20)
```

```python
temp['life_per'] = (temp['life exp']-temp['life exp'].shift(1))/temp['life exp'].shift(1)
temp['life_per'] = temp['life_per']*100
temp.head(20)
```
